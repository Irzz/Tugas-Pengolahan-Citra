```python
import PIL
from PIL import Image
 
img = Image.open("D:\Codingan\Python\Tugas Pengolahan Citra\R.jpg")
img = img.resize((700, 500))

#binarization function
def binarize(img):
  
    #initialize threshold
    thresh = 128

    #convert image to greyscale
    img = img.convert('L') 
    width, height = img.size
    
    #traverse through pixels 
    for x in range(width):
        for y in range(height):
        
            #if intensity less than threshold, assign white
            if img.getpixel((x,y)) < thresh:
                img.putpixel((x,y),0)

            #if intensity greater than threshold, assign black 
            else:
                img.putpixel((x,y),255)
            
    return img
```
